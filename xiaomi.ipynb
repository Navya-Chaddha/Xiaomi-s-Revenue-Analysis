```python
import pandas as pd 
import matplotlib.pyplot as plt

revenue = pd.read_csv("xiaomi_data.csv")
revenue
```

```python
revenue.iloc[1] # the first row containing the revenue of the company through smart phones 
```

```python
smartphones = pd.DataFrame({'Financial Year' :['FY2019','FY2020','FY2021','FY2022','FY2023','FY2024','FY2025'],'Revenue' : revenue.iloc[1,6:].values})
smartphones
```

```python
iot_lifestyle = pd.DataFrame({'Financial Year' :['FY2019','FY2020','FY2021','FY2022','FY2023','FY2024','FY2025'],'Revenue' : revenue.iloc[2,6:].values})
iot_lifestyle
```

```python
smartphones['Revenue'] = pd.to_numeric(smartphones['Revenue'])
smartphones.info()
```

```python
iot_lifestyle['Revenue'] = pd.to_numeric(iot_lifestyle['Revenue'])
iot_lifestyle.info()
```

```python
plt.plot(smartphones['Financial Year'],smartphones['Revenue'], marker = 'o')
plt.xlabel("Financial Year")
plt.ylabel("Revenue (CNY)")
plt.title('Smartphones')
plt.savefig('output/revenue_smartphone.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
#plotting the iot and lifestyle product data 
plt.plot(iot_lifestyle['Financial Year'],iot_lifestyle['Revenue'], color = 'black', marker = 'x')
plt.xlabel("Financial Year")
plt.ylabel("Revenue (CNY)")
plt.title('IOT and lifestyle product')
plt.savefig('output/revenue_iot_lifestyle.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
smartphones['Growth Percentage'] = smartphones['Revenue'].pct_change() * 100.0
smartphones
```

```python
iot_lifestyle['Growth Percentage'] = iot_lifestyle['Revenue'].pct_change()*100.0
iot_lifestyle
```

```python
plt.plot(smartphones['Financial Year'], smartphones['Growth Percentage'], color = 'red',label = 'smartphones')
plt.plot(smartphones['Financial Year'], iot_lifestyle['Growth Percentage'], color = 'blue',label = 'iot and lifestyle devices')
plt.xlabel('Financial Year')
plt.ylabel('Growth Percentage')
plt.title('Growth% vs financial year')
plt.legend()
plt.savefig('output/growth_vs_fy.png', dpi=150, bbox_inches='tight')
```

```python
st = smartphones['Revenue'].iloc[0]
end = smartphones['Revenue'].iloc[-1]
years = len(smartphones) - 1
s_cagr = ((end/st)**(1/years) - 1) * 100.0
s_cagr
```

```python
st2 = iot_lifestyle['Revenue'].iloc[0]
end2 = iot_lifestyle['Revenue'].iloc[-1]
years2 = len(iot_lifestyle) - 1
iot_cagr = ((end2/st2) ** (1/years2) - 1) *100.0
iot_cagr
```

```python
comparision = pd.DataFrame({'Financial Year' : smartphones['Financial Year'],'SmartPhone Revenue' : smartphones['Revenue'], 'IOT & Lifestyle Revenue' : iot_lifestyle['Revenue']})
comparision
```

```python
#Calculating the revenue share 
comparision['RS Smartphones'] = (comparision['SmartPhone Revenue'] / (comparision['SmartPhone Revenue']+ comparision['IOT & Lifestyle Revenue']))
comparision['RS IOT & Lifestyle'] = (comparision['IOT & Lifestyle Revenue'] / (comparision['SmartPhone Revenue']+ comparision['IOT & Lifestyle Revenue']))
comparision
```

```python
plt.figure(figsize=(8,5))
plt.plot(
    comparision['Financial Year'],
    comparision['RS Smartphones']*100,
    marker='o',
    label='Smartphones'
)

plt.plot(
    comparision['Financial Year'],
    comparision['RS IOT & Lifestyle']*100,
    marker='o',
    label='IoT & Lifestyle'
)

plt.ylabel('Revenue Share (%)')
plt.title('Revenue Composition of Xiaomi')
plt.legend()
plt.grid(True)
plt.savefig('output/revenue composition.png', dpi=150, bbox_inches='tight')
plt.show()
```
